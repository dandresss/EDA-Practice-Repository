```python
import pandas as pd
df = pd.read_csv('air+quality/AirQualityUCI.csv', sep=';')
df
```

```python
#Eliminar las 2 ultimas columnas y las filas donde todos los elementos son NaN
df = df.iloc[:,:-2].dropna(how='all')
df
```

```python
#Convertir en parquet. Ventajas: mas rapido, integridad de datos, menos espacio y apache sparl
df.to_parquet('clean_airquality.parquet')
```

```python
df.describe()
```

```python
#Agregar timestamp combinando dos filas
df['timestamp']= pd.to_datetime(df['Date'] + ' ' + df['Time'], format = '%d/%m/%Y %H.%M.%S')  
df['timestamp']
```

```python
#Agregarlo al dataframe df como primera columna y eliminar columnas Date y Time
df = df.drop(columns=['Date', 'Time'])
df = df[['timestamp']+df.columns[:-1].to_list()]
df
```

```python
df.dtypes
```

```python
#Obtenemos el tipo objeto debido a los numeros decimales con coma. 
#Debimos modificar nuestra lectura inicial del df usando la siguiente linea
#df = pd.read_csv('air+quality/AirQualityUCI.csv', sep=';')
#Ahora tendremos que usar un procedimiento distinto 
objects_cols= df.select_dtypes(include='object').columns.tolist()
objects_cols
```

```python
for column in objects_cols:
    df[column] = df[ column].str.replace(',','.').astype('float')
df.dtypes
```

```python
#Existen valores que aunque no son explicitamente NaN, denotan serlo con otros valores como -200
#Reemplazaremos los -200 con los valores que realmente significan
df.info()
```

```python
import numpy as np
df = df.replace(-200, np.nan)
df
```

```python
df.info()
#ahora podemos observar que muchos valores dependian de -200 para no ser expredos como nan 
```

```python
#Para poder definir el rango superior y el rango inferior, debemos usar
# La teoria de los cuartiles. Procederemos a realizar los calculos
Q1 = df['CO(GT)'].quantile(0.25)
Q3 = df['CO(GT)'].quantile(0.75)
IQR = Q3 - Q1
Lsup = Q3 + 1.5 * IQR
Linf = Q1 - 1.5 * IQR 
print(f'El rango superior es {Q1:.2f}')
print(f'El rango superior es {Q3:.2f}')
```

```python
#Aplicaremos nuestros resultados al dataframe y lo llamaremos inbounders
inbounders = df[(df['CO(GT)']<Lsup ) & (df['CO(GT)']>Linf)]
inbounders
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.histplot(df['CO(GT)'], bins = 30, kde = True)
plt.title('Histograma de CO(GT)')
plt.xlabel('Concentracion de CO(GT)')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha = 0.75)
plt.show()
```
